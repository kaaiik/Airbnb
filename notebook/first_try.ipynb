# Importaciones

```python
import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

# Carga de datos

```python
url = 'https://raw.githubusercontent.com/kaaiik/Airbnb-Predict/master/data/train.csv'
train = pd.read_csv(url, encoding='unicode_escape')
train.head(2)
```

```python
url2 = 'https://raw.githubusercontent.com/kaaiik/Airbnb-Predict/master/data/test.csv'
test = pd.read_csv(url2, encoding='unicode_escape')
test.head(2)
```

# Entrenamiento y test

```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
# parsear datos para h20

h2train=h2o.H2OFrame(train)
```

```python
X=[c for c in h2train.columns if c!='price']

y='price'
```

```python
# inicia auto-machine-learning

automl=H2OAutoML(max_models=50,
                 seed=42,   # random_state
                 max_runtime_secs=300,
                 sort_metric='RMSE')

```

```python
# entrena

automl.train(x=X,
             y=y,
             training_frame=h2train)
```

```python
print('[INFO] Leader board:')

leader_board=automl.leaderboard

leader_board.head()
```

```python
# prediciones del lider
h2test=h2o.H2OFrame(test)

y_pred=automl.leader.predict(h2test)
```

```python
lst_pred = h2o.as_list(y_pred)
lst_pred
```

```python
sample = pd.read_csv('https://raw.githubusercontent.com/kaaiik/Airbnb-Predict/master/data/sample.csv', encoding='unicode_escape')
sample.head()
```

```python
sample.price = lst_pred
```

```python
mod1 = sample
mod1.head()
```

```python
mod1.to_csv('prediccion.csv', index=False)
```
